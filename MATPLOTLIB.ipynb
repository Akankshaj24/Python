### Matplotlib - Used for visualization

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

##### **BarChart**
Used to compare continuous vallues in 2 different categories

```python
days = [1, 2, 3, 4, 5, 6, 7]
sales = [42, 38, 40, 45, 34, 39, 41]
print(len(days), len(sales))
```

```python
plt.bar(days, sales, color="orange", edgecolor="black", label="Sales")
plt.title("Day wise sales")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.xlim(0,9)
plt.ylim(0,60)
plt.legend()
plt.show()
```

### **Line chart**
 used to represent the trends

```python
# markers = o, *, <, >, 1, 2, 3, +, d, D
# for dashed dotted linestyle we can use --> '-.'
# lw --> line width

plt.plot(days, sales, color="green", label="Sales", marker="*",
         linestyle="dotted", lw=3, markersize=9)
plt.title("Day wise sales")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.xlim(0,9)
plt.ylim(0,60)
plt.legend()
plt.show()
```

#### **Scatter chart**
used to represent corelation values

**correlation** represents the magnitude and direction of linear relationship between 2 variables

```python
sales = np.random.randint(300,1000, 50)
profit = np.random.random(50)*np.random.randint(10,25,1)
print(sales)
print(profit)
```

```python
plt.scatter(sales, profit,color="maroon",marker="D")
plt.show()
```

#### **Histogram**
used to represent frequency distribution

```python
data = np.random.choice([1, 2, 3, 4, 5, 6, 7, 8, 9], size = 20)
print(data)
```

```python
from collections import Counter
```

```python
c1 = Counter(data)
print(c1)
```

```python
plt.hist(data, edgecolor="black", color="lightblue")
plt.grid()
plt.show()
```

```python
print(c1)
plt.hist(data, bins="auto")
plt.show()
```

```python
print(c1)
plt.hist(data, bins=9)
plt.show()
```

#### **Pie chart**
used to represent percentage distribution

```python
marks = [78, 89, 65, 72, 80]
sub = ["Python", "SQL", "Tableau", "ML", "Excel"]
```

```python
plt.pie(marks, labels=sub, autopct="%.2f%%")
plt.show()
```

#### BOX-PLOT
1. Used to detect outliers
2. Used present 5 points summary

**5 Points** -->
1. Q1 : 25 percentile
2. Q2 : 50 percentile
3. Q3 : 75 percentile
4. min : Q1 - 1.5*IQR
5. max : Q3 + 1.5*IQR
6. IQR : Q3-Q1 (Inter Quartile Range)
7. Outliers : data points > max or data points < min

```python
w = [14,12,18,19,21,40,27,26,25,14,3,22,24,20,19,15,14,20,39,13,12,11,10,15,17,18,19,16,17,18,19]
print(w)
```

```python
plt.boxplot(w)
plt.show()
```
